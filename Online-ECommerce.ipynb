```python
import seaborn as sns
df = pd.read_csv(r"C:\Users\Admin\OneDrive\Documents\Dataset\Online-eCommerce.csv")

plt.figure(figsize=(12, 6))
sns.countplot(x='Category', data=df,hue='Category',palette='CMRmap')
plt.title('Count of Entries in Each Category')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=40)
plt.show()
```

```python
category_sales = df.groupby('Category')['Sales'].sum()
category_sales.plot(kind='bar', color='skyblue', figsize=(12, 6))
plt.title('Total Sales by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=40)
plt.show()
```

```python
plt.figure(figsize=(12, 6))

plt.hist(df['Sales'], bins=10, color='skyblue', edgecolor='black',linestyle='--')
plt.grid(axis='y', linestyle='--', linewidth=0.5)
plt.title('Distribution of Sales')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()
```
